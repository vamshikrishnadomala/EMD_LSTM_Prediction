```python
from PyEMD import EMD
import numpy as np
import pylab as plt
```

```python
import pandas as pd
df=pd.read_excel(r"51000_colab.xlsx", engine='openpyxl')
df.head()
```

```python
plt.plot(df['Date'],df['WVHT'])
```

```python
df.shape
```

```python
s=df[['WVHT']].values
```

```python
s
```

```python
s=s.ravel()
```

```python
s.shape
```

```python
# Define signal
t = np.linspace(0, 1, 52608)
# Execute EMD on signal
IMF = EMD().emd(s,t)
N = IMF.shape[0]+1
#df=pd.DataFrame(,IMF)
#df.to_excel('imf.xlsx')
```

```python
IMF
```

```python
IMF2=IMF.T
```

```python
IMF2
```

```python
IMF2.shape
```

```python
IMF.shape
```

```python
df2=pd.DataFrame(IMF2, columns=['imf1','imf2','imf3','imf4','imf5','imf6','imf7','imf8','imf9','imf10',
                              'imf11','imf12','imf13','imf14','imf15'])
```

```python
df2
```

```python
df2.to_excel('IMF_DF2.xlsx')
```

```python
df3= pd.DataFrame(data = IMF2[0:, 0:],
                        index = ['Row_' + str(i + 1) 
                        for i in range(IMF2.shape[0])],
                        columns = ['imf_' + str(i + 1) 
                        for i in range(IMF2.shape[1])])
  
# printing the dataframe
print(df3)
```

```python
df2.to_excel('IMF_DF3.xlsx')
```

```python
# Plot results
plt.rcParams['figure.figsize'] = [12, 9]
plt.subplot(N,1,1)
plt.plot(t, s, 'r')
plt.title("Input signal: Wave Height")
plt.xlabel("Time [s]")
for n, imf in enumerate(IMF):
  plt.subplot(N,1,n+2)
  plt.plot(t, imf, 'g')
  plt.title("IMF "+str(n+1))
  plt.xlabel("Time [s]")
  plt.subplots_adjust(left=0.1,
                    bottom=0.01, 
                    right=0.9, 
                    top=4, 
                    wspace=0.4, 
                    hspace=0.4)

#plt.tight_layout()
plt.savefig('simple_example')
plt.show()
#df=pd.DataFrame(imf1)
#df.to_excel('imf.xlsx')
```
